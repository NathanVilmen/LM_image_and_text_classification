# Inference Script

## Load in Packages 
Likely that most of these arent needed

```python
from torchvision import transforms
from PIL import Image
from torchvision import models
import torch.nn as nn
import torch
import pandas as pd
import numpy as np
import os
# os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
```

## Load in the Model

```python
model = models.resnet50()

# # Change the last layer to classify into 6 classes
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 6)

# Define the file paths
model_path = 'model_weights.pth'
print(model_path)

# Load the saved model
state_dict = torch.load(model_path)
model.load_state_dict(state_dict)
model.eval()
```

# Run the Inference Script 
Loads in the weights of the saved model and runs the model for each image in the folder.

```python
# Define the transformations to be applied to each image
transform = transforms.Compose([
    transforms.Resize((112, 112)),
    transforms.ToTensor()
])
print(['buildings(0)', 'forest(1)', 'glacier(2)', 'mountain(3)', 'sea(4)', 'street(5)'])
# Loop through all images in the folder
outputs = []
for filename in os.listdir():
    if filename.endswith('.jpg'):
        # Open and transform the image
        img = Image.open(filename)
        img = transform(img)

        # Make a prediction
        with torch.no_grad():
            output = model(img.unsqueeze(0))
            prediction = torch.argmax(output, dim=1)
        
        # Print the filename and predicted class
        print(f'{filename}: {prediction.item()}')
        row = [filename, prediction.item()]
        outputs.append(row)

np.array(outputs)

data = pd.DataFrame(outputs)
data.to_csv('preds.csv', header=['image', 'class'], index=False)
```
